```python
import logging
logging.basicConfig(filename="5MarchInfo.log", level=logging.INFO, format="%(asctime)s %(name)s %(message)s")
     

# answer 1
# plotting graph using bokeh python library
from bokeh.plotting import figure, output_file,show

logging.info("data for the plottig the graph")
x= [2,8,1,6]
y = [4,8,9,3]

p = figure(title = " Bokeh line plot")
p.title.text_font_size = "3em"
p.title.text_color = "green"
logging.info("labelling x and y axis")
p.xaxis.axis_label = "xData"
p.yaxis.axis_label= "YData"

logging.info("plotting line graph")
p.line(x,y, legend_label = "Line", line_width= 3,line_color= 'blue' )
show(p)
```

```python
# answer 2
# glyphs
"""
In Bokeh, glyphs are the basic visual building blocks for creating interactive visualizations. 
Glyphs are geometric shapes that can be plotted at specific locations in a plot and can have different
visual properties such as size, color, and transparency. Bokeh provides several different types of glyphs,
including circles, squares, lines, and more.
"""
# Example of how to add a circle glyph  

# Prepare the data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

logging.info("specifying the output file")
output_file("circle.html")

logging.info("creating a new plot")
p = figure(title="Circle Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

logging.info("Add a circle glyph with the x and y data")
p.circle(x, y, size=15, color="green", alpha=0.5)

# Show the plot
show(p)
     
```

```python
# answer 3
"""
We can customize the appearance of a plot in many ways, including the axes, title, legend, and more.
"""
# Example:
from bokeh.plotting import figure, output_file, show
from bokeh.models import NumeralTickFormatter

logging.info("preparing the data")
x = [11, 12, 3, 14, 5]
y = [6, 17, 2, 14, 15]

logging.info("specifying the output file")
output_file("custom.html")

logging.info("creating a new plot")
p = figure(title="Custom Plot", x_axis_label='X-axis', y_axis_label='Y-axis', 
           plot_width=500, plot_height=400)

logging.info("")
# Add a line glyph with the x and y data
p.line(x, y, legend_label="Line", line_width=2)

logging.info("")
# Customize the appearance of the plot

logging.info("title axes and legends")
p.title.text_color = "purple"
p.title.text_font = "italic"
p.title.text_font_size = "5em"
p.xaxis.axis_label_text_font = "arial"
p.yaxis.axis_label_text_font = "italic"
p.xaxis.axis_label_text_font_style = "italic"
p.yaxis.axis_label_text_font_style = "italic"
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_color = "red"
p.xaxis.major_label_text_color = "green"
p.yaxis.major_label_text_color = "green"
p.xaxis.formatter = NumeralTickFormatter(format="0.00")
p.legend.location = "top_left"
p.legend.label_text_font_size = "14pt"
p.legend.background_fill_color = "beige"
p.legend.border_line_color = None
show(p)
```
