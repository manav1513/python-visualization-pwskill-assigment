```python
# Q1: Five plots that can be plotted using Seaborn library:

# Scatter plot - It is used to visualize the relationship between two continuous
# variables.

# Line plot - It is used to show the trend of a variable over time or across 
#             different categories.
# Bar plot - It is used to compare values across different categories using 
#             vertical or horizontal bars.
# Heatmap - It is used to show the relationship between two variables using 
#             color-coded cells.
# Pair plot - It is used to visualize the relationship between multiple variables
#             in a dataset.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sn
sn.load_dataset('fmri').head()
```

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot a line plot with timepoint on the x-axis and signal on the y-axis
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue
# parameter for the "species" column
# of the iris dataset.
```

```python
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot a pair plot of the iris dataset with species as the hue parameter
sns.pairplot(data=iris, hue="species")
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset("flights")

# Reshape the data to be in the form of a pivot table
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot a heatmap of the pivot table
sns.heatmap(data=flights_pivot, cmap="coolwarm")
```
