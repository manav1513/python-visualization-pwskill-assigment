```python
# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
# Matplotlib

# ans = Matplotlib is a Python library for creating static, animated, and 
# interactive visualizations in Python. It is one of the most widely used 
# plotting libraries in the scientific Python community.

# Matplotlib is used for creating visualizations to represent data in a more
# understandable and graphical way. It provides a wide range of plot types and
# customization options that can help users create high-quality and 
# publication-ready graphs for their research or projects.

# The Pyplot module is a sub-library of Matplotlib that provides a simple
# interface for creating various types of plots. Some of the plots that can be
# plotted using the Pyplot module of Matplotlib include:

# Line plot: A plot that displays data points connected by straight lines. It is
# commonly used to show trends in data over time.

# Bar plot: A plot that represents data with rectangular bars. It is often used
# to compare different categories of data.

# Scatter plot: A plot that shows the relationship between two variables by
# plotting their values as points on a two-dimensional plane.

# Histogram: A plot that shows the distribution of data in a dataset. It is
# commonly used to visualize the frequency of values in a dataset.

# Pie chart: A plot that shows the proportion of different categories in a 
# dataset as slices of a pie. It is often used to show how a whole is divided 
# into parts.
```

```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.

# ans = A scatter plot is a diagram where each value in the data set is represented
# by a dot. The Matplotlib module has a method for drawing scatter plots, it needs 
# two arrays of the same length, one for the values of the x-axis, and one for the 
# values of the y-axis

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
```

```python
plt.sunplot(x,y)
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the 
# subplot() function.
# Use the following data:
    
# ans = The subplot() function takes three arguments that describes the layout 
# of the figure. The layout is organized in rows and columns, which are represented
# by the first and second argument.
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create the subplots
fig, axs = plt.subplots(nrows=2, ncols=2)

# Plot the data on each subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar 
# plot and a horizontal bar plot.

# ans = A bar plot is a chart that displays data with rectangular bars, where the
# length of each bar is proportional to the value of the data it represents. 
# Bar plots are commonly used to compare different categories of data and to
# show changes in data over time. They can be created vertically or horizontally,
# depending on the orientation of the bars.

# Bar plots are used because they make it easy to compare the values of different
# categories or groups of data. They are also useful for showing changes in data
# over time, especially when combined with other visualization techniques like
# stacked bar plots or grouped bar plots.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.barh(company, profit)
# Add labels and a title
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits')

# Show the plot
plt.show()
```

```python
plt.bar(company, profit)
plt.show()

# Add labels and a title
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits')
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box 
# plot.

# ans = A box plot is a chart that displays the distribution of data in a dataset
# by showing the median, quartiles, and outliers of the data. It is also known 
# as a box-and-whisker plot. Box plots are commonly used to compare the 
# distribution of data between different groups or to identify outliers in a 
# dataset.

# Box plots are used because they provide a visual summary of the distribution of
# data in a dataset, including information about the median, quartiles, and
# outliers. They are useful for identifying skewness, outliers, and the spread of
# the data.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
```

```python
# Create the box plot
plt.boxplot(data)

# Add labels and a title
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```
